```python
import rioxarray as rio
import xarray as xr
```

# Read green and SWIR band 

```python
fn_green = './HLS_data/T10TES/HLS.L30.T10TES.2021014T185549.v2.0/HLS.L30.T10TES.2021014T185549.v2.0.B03.tif'
fn_swir = './HLS_data/T10TES/HLS.L30.T10TES.2021014T185549.v2.0/HLS.L30.T10TES.2021014T185549.v2.0.B11.tif'
hls_green = rio.open_rasterio(fn_green, chuncks=True)
hls_swir = rio.open_rasterio(fn_swir, chuncks=True)

```

```python
hls_green.plot()
```

```python
ndsi = (hls_green-hls_swir)/(hls_green+hls_swir) 
```

```python
sca = xr.where(ndsi > 0.4, 1, 0)
sca.plot()
```
